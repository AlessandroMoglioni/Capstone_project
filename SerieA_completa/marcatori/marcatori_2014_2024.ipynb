```python
import pandas as pd

# Elenco dei percorsi completi ai CSV (modifica con i tuoi)
file_paths = [
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA1415\marcatori_serieA1415.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA1516\marcatori_serieA1516.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA1617\marcatori_serieA1617.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA1718\marcatori_serieA1718.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA1819\marcatori_serieA1819.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA1920\marcatori_serieA1920.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA2021\marcatori_serieA2021.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA2122\marcatori_serieA2122.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA2223\marcatori_serieA2223.csv",
    r"C:\Users\Alessandro\Desktop\Progetto_epicod\Partite_serieA\SerieA2324\marcatori_serieA2324.csv",
]

# Lista per contenere i DataFrame
df_list = []

for path in file_paths:
    df = pd.read_csv(path)
    
    # Rendi i nomi delle colonne coerenti
    df.columns = df.columns.str.strip().str.lower()
    
    # Rinomina colonne in modo uniforme (opzionale ma consigliato)
    df = df.rename(columns={
        'gol totali': 'gol_totali',  # nome unico standard
        'giocatore': 'giocatore',
        'squadra': 'squadra',
        'anno': 'anno'
    })
    
    df_list.append(df)

# Unione dei DataFrame
merged_df = pd.concat(df_list, ignore_index=True)

# Salvataggio
merged_df.to_csv('marcatori_2014_2024.csv', index=False)
```
